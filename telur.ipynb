```python
# mengimportkan gambar telur nya(fungsi menampilkan gambar)
from IPython.display import Image
import numpy as np # library untuk mengolah matriksnya
import cv2 #cv2 untuk mengelola gambar
Image(filename='kotor.jpg') 
```

```python
image = cv2.imread("kotor.jpg") #mengambil gambar dijadikan bentuk array dengan format bgr
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
import matplotlib.pyplot as plt 
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.show() #menampilkan gambar
for i in range(len(img_gray)): # menampilkan data pixel
    for j in range(len(img_gray[0])):
        print(str(img_gray[i][j])+",", end="")
    print()
```

```python
# for j in range(len(numpy_array[0])):
#             if numpy_array[i][j] > threshold:
#                 numpy_array[i][j] = 255
#             else:
#                 numpy_array[i][j] = 0
#     return numpy_array

for i in range(len(img_gray)): #mengubah warna menjadi hitam putih 
    for j in range(len(img_gray[0])):
        if img_gray[i][j] > 100:
            img_gray[i][j] = 255
        else:
                img_gray[i][j] = 0
#     return img_array
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.show() #menampilkan gambar
```

```python
# menghitung jumlah pixel warna hitam 
hitung=0
for i in range(len(img_gray)):
    for j in range(len(img_gray[0])):
        if img_gray[i][j] == 0:
            hitung=hitung+1
#     return img_array
if hitung > 1000:
    print('telur kotor')
else:
    print('telur bersih')
```
